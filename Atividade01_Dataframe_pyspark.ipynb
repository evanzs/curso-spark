```python
## resolvendo primeira questão
# 1.Crie uma consulta que mostre,nesta ordem, Nome, Estados e status

dfs_cliente.select('Cliente','Estado','Genero').show(5)
```

```python
# 2.Crie uma consulta que mostre apenas os clientes do Status “platinum” e “gold”

dfs_cliente.filter((dfs_cliente.Status == 'Gold') | (dfs_cliente.Status == 'Platinum')).collect()
```

```python
# 2.Crie uma consulta que mostre apenas os clientes do Status “platinum” e “gold”

dfs_cliente.where((dfs_cliente.Status == 'Platinum') | (dfs_cliente.Status == 'Gold')).collect()
```

```python
dfs_vendas.where((dfs_vendas.ClienteID == '54')).collect()
```

```python
# 3.Demostre quanto cada Status de Clientes representa em vendas?

df_cliente_vendas = dfs_vendas.join(dfs_cliente,'ClienteID').select('VendasID','Cliente','Total','Status') ## join pelo ClienteID e selecionando tabelas pertinentes
```

```python
from pyspark.sql.types import FloatType
from pyspark.sql import functions as f
df_cliente_vendas.withColumn('Total',df_cliente_vendas.Total.cast(FloatType()))
df_cliente_vendas.groupBy('Status').agg(f.sum('Total')).show(5)
```
